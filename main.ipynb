```python
# load the dataset
import pandas as pd
covid = pd.read_csv("EPL20-21.csv")
non_covid = pd.read_csv("EPL24-25.csv")
```

```python
# filter team names and fulltime goals of each team
covid = covid[['HomeTeam','AwayTeam','FTHG','FTAG']]
non_covid = non_covid[['HomeTeam','AwayTeam','FTHG','FTAG']]
```

```python
covid.head(), non_covid.head()
```

```python
# calculate the home-away difference of each season
covid_sample = covid.sample(n=35, random_state=10).copy()
non_covid_sample = non_covid.sample(n=35, random_state=10).copy()
covid_sample["net_different"] = covid_sample["FTHG"] - covid_sample["FTAG"]
non_covid_sample["net_different"] = non_covid_sample["FTHG"] - non_covid_sample["FTAG"]
```

```python
covid_sample
```

```python
non_covid_sample
```

```python
# ttest via scipy.stats library
from scipy.stats import ttest_ind

# two-sample t-test
t_stat, p_value = ttest_ind(
    non_covid_sample["net_different"],
    covid_sample["net_different"],
    equal_var=False,
    alternative="two-sided"
)

print("Covid mean:", covid_sample["net_different"].mean())
print("Non-covid mean:", non_covid_sample["net_different"].mean())
print("Covid std:", covid_sample["net_different"].std())
print("Non-covid std:", non_covid_sample["net_different"].std())
print("t statistic:", t_stat)
print("p-value:", p_value)
```

```python
# interpretation
# Assuming the null hypothesis is true, and there is no difference in home-away mean differences between the two seasons,
# there is a probability of 0.4249 that we would get a sample difference in means where 
# mean home-away difference in non-covid season is greater OR less the mean home-away difference in covid season,
# purely by chance
```
